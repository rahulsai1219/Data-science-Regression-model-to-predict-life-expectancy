### From the above VFI values it can be incurred that income_composition_of_resources and schooling have VFI value more than 10 as these independent variables are corelating with each other

Hence removing one of these two to have VFI value less than 10

```python
x=full_data[['adult_mortality','income_composition_of_resources']]
y=full_data[['life_expectancy']]
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=1)

vif=pd.DataFrame()
vif["VIF"]=[variance_inflation_factor(x_train.values,i) for i in range(x_train.shape[1])]
vif["Features"]=x_train.columns
vif
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg=LinearRegression()
```

```python
print(linreg.intercept_)
```
